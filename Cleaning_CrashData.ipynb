### Filling NaN values in different columns:

**Here, the two features 'Related Non-Motorist' and 'Non-Motorist Substance Abuse' have almost 80,000+ null values. Assuming that when there is no Non-Motorist involved in the crash, these values are null. So instead of leaving it null, I am filling with 'NONE' for those where both the features are null at the same time (new category when there is no non motorist involved)**

```python
null_rows = df_road[df_road['Related Non-Motorist'].isna() & df_road['Non-Motorist Substance Abuse'].isna()]
null_rows[['Related Non-Motorist', 'Non-Motorist Substance Abuse']] = null_rows[['Related Non-Motorist', 'Non-Motorist Substance Abuse']].fillna('NONE')
df_road.update(null_rows)
```

```python
df_road['Lane Type'].fillna('UNKNOWN', inplace=True)
df_road['Municipality'].fillna('UNKNOWN', inplace=True)
df_road['Second Harmful Event'].fillna("NONE", inplace=True)
df_road['Fixed Oject Struck'].fillna("NONE", inplace=True)
```

```python
null_rows = df_road[df_road['Intersection Area'].isna() & df_road['Intersection Type'].isna()]
null_rows[['Intersection Area', 'Intersection Type']] = null_rows[['Intersection Area', 'Intersection Type']].fillna('NONE')
df_road.update(null_rows)
```

```python
df_road["Intersection Type"].fillna("UNKNOWN", inplace=True)
df_road["Intersection Area"].fillna("UNKNOWN", inplace=True)
df_road['Junction'].fillna('UNKNOWN', inplace=True)
df_road['Traffic Control'].fillna('UNKNOWN', inplace=True)
df_road['Driver Substance Abuse'].fillna('UNKNOWN',inplace=True)
df_road['Weather'].fillna('UNKNOWN',inplace=True)
df_road['Road Condition'].fillna('UNKNOWN', inplace=True)
df_road['Surface Condition'].fillna('UNKNOWN', inplace=True)
df_road['Road Division'].fillna('UNKNOWN', inplace=True)
df_road['Route Type'].fillna('UNKNOWN',inplace=True)
df_road['Mile Point Direction'].fillna('Unknown',inplace=True)
df_road['Road Grade'].fillna('UNKNOWN',inplace=True)
df_road['Cross-Street Type'].fillna('Unknown',inplace=True)
df_road['Collision Type'].fillna('UNKNOWN',inplace=True)
df_road['Light'].fillna('UNKNOWN',inplace=True)
df_road['Non-Motorist Substance Abuse'].fillna('UNKNOWN',inplace=True)
df_road['First Harmful Event'].fillna('UNKNOWN',inplace=True)
df_road['Road Alignment'].fillna('UNKNOWN',inplace=True)
df_road['Direction'].fillna('Unknown', inplace=True)
```

```python
df_road['Hit/Run'].fillna('No',inplace=True)
```

```python
numeric_probabilities = df_road['Distance'].value_counts(normalize=True)

missing_indices = df_road[df_road['Distance'].isnull()].index

df_road.loc[missing_indices, 'Distance'] = np.random.choice(
    numeric_probabilities.index,
    size=len(missing_indices),
    p=numeric_probabilities.values
)
```

```python
numeric_probabilities = df_road['Mile Point'].value_counts(normalize=True)

missing_indices = df_road[df_road['Mile Point'].isnull()].index

df_road.loc[missing_indices, 'Mile Point'] = np.random.choice(
    numeric_probabilities.index,
    size=len(missing_indices),
    p=numeric_probabilities.values
)
```
